```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
from prophet import Prophet
# from lstr import LSTR
```

```python
# Split data into training and testing sets
train = df[:1000]
test = df[1000:]
```

```python
# Preprocess data
scaler = StandardScaler()
X_train = scaler.fit_transform(train.drop(['date', 'close'], axis=1))
y_train = train['close']
X_test = scaler.transform(test.drop(['date', 'close'], axis=1))
y_test = test['close']
```

```python
# Random Forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
mae_rf = np.mean(np.abs(y_pred_rf - y_test))
mse_rf = np.mean(np.square(y_pred_rf - y_test))
```

```python
# Support Vector
svm = SVR()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
mae_svm = np.mean(np.abs(y_pred_svm - y_test))
mse_svm = np.mean(np.square(y_pred_svm - y_test))
```

```python
# Gradient Boosting
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
mae_gb = np.mean(np.abs(y_pred_gb - y_test))
mse_gb = np.mean(np.square(y_pred_gb - y_test))
```

```python
# Prophet
prophet_train = train[['date', 'close']].rename(columns={'date': 'ds', 'close': 'y'})
prophet_test = test[['date', 'close']].rename(columns={'date': 'ds', 'close': 'y'})
prophet = Prophet()
prophet.fit(prophet_train)
forecast = prophet.predict(prophet_test)
y_pred_prophet = forecast['yhat'].values
mae_prophet = np.mean(np.abs(y_pred_prophet - y_test))
mse_prophet = np.mean(np.square(y_pred_prophet - y_test))
```

```python
def clean_data(data, ticker_name):
    # select only 'Open' and 'Close' columns
    data = data[['Open', 'Close']].copy()

    # rename columns with ticker symbol
    data.columns = [f"{ticker_name} {col}" for col in data.columns]

    # set index as a datetime object
    data.index = pd.to_datetime(data.index)

    # drop columns before 2019-08-29 and after 2023-03-08
    data = data.loc['2019-08-29':'2023-03-08']

    return data
```

```python
import yfinance as yf
import pandas as pd

# Define dictionary of ticker_name : ticker
tickers = {'Nikkei 225': '^N225', 
           'S&P 500': '^GSPC', 
           'CSI 300': '000300.SS', 
           'US Dollar Index': 'DX-Y.NYB', 
           'Dow Jones Industrial Average': '^DJI', 
           'Crude Oil': 'CL=F', 
           'NASDAQ Composite': '^IXIC'}

# Retrieve and clean data for each ticker
dfs = []
for ticker_name, ticker in tickers.items():
    # Retrieve data
    data = yf.Ticker(ticker).history(period='5y', interval='1d')
    # Clean data
    data = clean_data(data, ticker_name)
    # Append to list of data frames
    dfs.append(data)

```
