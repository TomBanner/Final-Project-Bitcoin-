```python
import requests
import pandas as pd

# Set the start and end dates for the time period you want
start_date = '2019-08-28'
end_date = '2023-03-08'

# Define the symbols for the assets you want to retrieve
symbols = {
    'P225': '^N225',
    'S&P500': '^GSPC',
    'CSI300': '000300.SS',
    'DXY': 'DX-Y.NYB',
    'DJI': '^DJI',
    'Oil': 'CL=F',
    'NASDAQ': '^IXIC'
}

# Create an empty dataframe to store the data
prices = pd.DataFrame()

# Loop through each symbol and append the data to the dataframe
for symbol in symbols.values():
    url = "https://seeking-alpha.p.rapidapi.com/market/get-realtime-quotes"
    querystring = {"sa_ids": symbol}
    headers = {
        "X-RapidAPI-Key": "40d3a7f43cmshb0c6326b8f21b2cp1f3f76jsn6989f38dacf4",
        "X-RapidAPI-Host": "seeking-alpha.p.rapidapi.com"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    data = response.json()['data'][0]
    df = pd.DataFrame(data, index=[pd.to_datetime(data['time'])])
    df = df[['last_price', 'volume']]
    df.columns = [f'{symbol}_{col}' for col in df.columns]
    prices = pd.concat([prices, df], axis=1)

# Filter the dataframe by date range
prices = prices[(prices.index >= start_date) & (prices.index <= end_date)]

# Sort the dataframe by date
prices = prices.sort_index()

# Print the first few rows of the dataframe
print(prices.head())
```
