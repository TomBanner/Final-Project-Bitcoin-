**Getting other financial data from YFinance to be used as features**

```python
#Getting Data from Yfinance

# Set the start and end dates
start_date = '2019-08-29'
end_date = '2023-03-08'

# Get the S&P 500 daily performance data
sp500_df = yf.download('^GSPC', start=start_date, end=end_date, progress=False)
sp500_daily_performance = (sp500_df['Close'] - sp500_df['Open']) / sp500_df['Open'] * 100

# Get the NASDAQ 100 daily performance data
nasdaq100_df = yf.download('^NDX', start=start_date, end=end_date, progress=False)
nasdaq100_daily_performance = (nasdaq100_df['Close'] - nasdaq100_df['Open']) / nasdaq100_df['Open'] * 100

# Get the treasury bond rate data
treasury_bond_rates_df = yf.download('^TNX', start=start_date, end=end_date, progress=False)
treasury_5yr_bond_rate = treasury_bond_rates_df['Close'].resample('D').last().ffill()
treasury_10yr_bond_rate = treasury_bond_rates_df['Close'].resample('D').last().ffill()
treasury_30yr_bond_rate = treasury_bond_rates_df['Close'].resample('D').last().ffill()

# Get the 1-year treasury bond rate data
treasury_1yr_bond_rates_df = yf.download('^IRX', start=start_date, end=end_date, progress=False)
treasury_1yr_bond_rate = treasury_1yr_bond_rates_df['Close'].resample('D').last().ffill()

# Get the bitcoin futures data
bitcoin_futures_df = yf.download('BTC=F', start=start_date, end=end_date, progress=False)
bitcoin_futures_volume = bitcoin_futures_df['Volume']
bitcoin_futures_daily_returns = (bitcoin_futures_df['Close'] - bitcoin_futures_df['Open']) / bitcoin_futures_df['Open'] * 100
bitcoin_futures_volatility = bitcoin_futures_daily_returns.std()

# Combine the data into a single DataFrame
data = {'S&P 500 Daily Performance': sp500_daily_performance,
        'NASDAQ 100 Daily Performance': nasdaq100_daily_performance,
        '1-Year Treasury Bond Rate': treasury_1yr_bond_rate,
        '5-Year Treasury Bond Rate': treasury_5yr_bond_rate,
        '10-Year Treasury Bond Rate': treasury_10yr_bond_rate, 
        '30-Year Treasury Bond Rate': treasury_30yr_bond_rate,
        'Bitcoin Futures Volume' : bitcoin_futures_volume, 
        'Bitcoin Futures Returns' : bitcoin_futures_daily_returns,
        'Bitcoin Futures Volatility': bitcoin_futures_volatility}

yfinance_data = pd.DataFrame(data) 
```

```python
#Dropping Volatility column 
yfinance_data.drop(columns=['Bitcoin Futures Volatility'], inplace=True)
```

**Getting data about economic/finacial indices**

```python
#Function to fill in the missing dates (eg weekends/holidays)

def fill_missing_values(df_name, start_date, end_date):
    # create a DatetimeIndex with every day from start date to end date
    idx = pd.date_range(start=start_date, end=end_date, freq='D')

    # create a DataFrame with the DatetimeIndex
    df = pd.DataFrame(index=idx)
    
    df_name.index = pd.DatetimeIndex(df_name.index).tz_localize(None)

    # merge the original DataFrame with the one with the DatetimeIndex
    df = df.merge(df_name, how='left', left_index=True, right_index=True)

    # fill missing values with previous day's values
    df.fillna(method='ffill', inplace=True)

    return df
```

```python
#Function to clean and reformat the dataframes

def clean_data(data, ticker_name):
    # select only 'Open' and 'Close' columns
    data = data[['Open', 'Close']].copy()

    # rename columns with ticker symbol
    data.columns = [f"{ticker_name} {col}" for col in data.columns]

    # set index as a datetime object
    data.index = pd.to_datetime(data.index)

    # drop columns before 2019-08-29 and after 2023-03-08
    data = data.loc['2019-08-29':'2023-03-08']

    return data
```

```python
#Retrieving data about indices from Yfinance

# Define dictionary of ticker_name : ticker
tickers = {'Nikkei 225': '^N225',
           'S&P 500': '^GSPC',
           'US Dollar Index': 'DX-Y.NYB',
           'Dow Jones Industrial Average': '^DJI',
           'Crude Oil': 'CL=F',
           'NASDAQ Composite': '^IXIC'}

# Retrieve and clean data for each ticker
dfs = []
for ticker_name, ticker in tickers.items():
    # Retrieve data
    data = yf.Ticker(ticker).history(period='5y', interval='1d')

    # Clean data
    data = clean_data(data, ticker_name)

    # Fill missing values
    data = fill_missing_values(data, '2019-08-29', '2023-03-08')

    # Append to list of data frames
    dfs.append(data)
```

```python
indices_data = pd.concat(dfs, axis=1)
```

```python
indices_data.to_csv('./data/indices_data.csv', index=True)
```
